# Neural network module (nn) from PyTorch.


```python
import torch
import torch.nn as nn
```

# Single Conv2d layer

```python
conv = nn.Conv2d(
    in_channels=3,
    out_channels=8,
    kernel_size=3,
    padding=1
)
```

# Activation Function

```python
relu = nn.ReLU()
```

Dummy Image
# (batch, channels, height, width)

```python
x = torch.randn(1,3,64,64)
print(x.shape)
```

```python
x = conv(x)
print(x.shape)
```

```python
x = relu(x)
print(x.shape)
```
